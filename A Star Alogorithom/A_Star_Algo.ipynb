```python
from  collections import deque
input_f=open("/content/drive/MyDrive/Courses ü§¶‚Äç‚ôÇÔ∏è/CSE 422/Lab_1/input.txt","r")
output_f=open("/content/drive/MyDrive/Courses ü§¶‚Äç‚ôÇÔ∏è/CSE 422/Lab_1/output.txt","a")

heuristic_value={}
paths={}

for i in range(20):

  initial=0
  input_set=list((input_f.readline().split()))

  while initial<len(input_set):
    if initial == 0:
      paths[input_set[initial]] = []
      heuristic_value[input_set[initial]] = int(input_set[initial+1])
      initial +=2

    else:
      paths[input_set[0]].append((input_set[initial], int(input_set[initial+1])))
      initial +=2

# print(heuristic_value)
# print(paths)


class a_star:

  def __init__(self,paths,heuristic_value):
    self.paths=paths
    self.heuristic_value=heuristic_value

  def heuristic(self,f):
    h=self.heuristic_value
    return h[f]

  def connected_cities(self,city_name):
    return self.paths[city_name]

  def main(self,start,ends):

    st=set([start])
    end=set([])

    parent_node={}
    parent_node[start] = start

    present_node={}
    present_node[start] =0

    while st:

      first = None

      for city in st:
        if first==None or present_node[city] + self.heuristic(city) < present_node[first] + self.heuristic(first):
          first = city

      if first == None:
        return None

      if first == ends:
        fianl_destination = []

        while parent_node[first] != first:
            fianl_destination.append(first)
            first = parent_node[first]

        fianl_destination.append(start)

        fianl_destination.reverse()

        cost = 0
        x = 0

        while x < (len(fianl_destination) - 1):
            for d,val in self.paths.items():
                if d == fianl_destination[x]:
                    for a,b in val:

                        if a == fianl_destination[x+1]:
                            cost += b
                            x += 1
                            break
                    break
        the_path = "Paths: "
        for i in fianl_destination:
            if i == fianl_destination[-1]:
                the_path = the_path + i
            else:
                the_path = the_path + i + " -> "

        output_f.write(f"{the_path}\nTotal distance: {cost} km\n\n")
        print(f"{the_path}\nTotal distance: {cost} km\n\n")
        return None




      for next,value in self.connected_cities(first):

        if next not in st and next not in end:
          st.add(next)
          parent_node[next]= first
          present_node[next]=present_node[first] + value

        else:
          if present_node[next]> present_node[first] + value:
            present_node[next]= present_node[first] + value
            parent_node[next]= first

            if next in end:
              end.remove(next)
              st.add(next)

      st.remove(first)
      end.add(first)

    output_f.write("Path does not exist!")
    # print("Path does not exist!")
    return None

star= a_star(paths,heuristic_value)
st=input("Start Node: ")
end=input("Dest Node: ")
star.main(st,end)
```
